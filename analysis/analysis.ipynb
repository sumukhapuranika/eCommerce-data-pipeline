```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

conn = sqlite3.connect("../db/ecommerce.db")
```

```python
query = """
SELECT order_month, SUM(revenue) as revenue
FROM ecommerce_data
GROUP BY order_month
ORDER BY order_month;
"""

df = pd.read_sql(query, conn)

plt.figure()
plt.plot(df['order_month'], df['revenue'])
plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend")
plt.show()
```

```python
query = """
SELECT customer_unique_id, SUM(revenue) as total_spent
FROM ecommerce_data
GROUP BY customer_unique_id
ORDER BY total_spent DESC
LIMIT 10;
"""

df = pd.read_sql(query, conn)

plt.figure()
plt.bar(df['customer_unique_id'], df['total_spent'])
plt.xticks(rotation=90)
plt.title("Top 10 Customers")
plt.show()
```

```python
query = """
SELECT customer_city, COUNT(*) as total_orders
FROM ecommerce_data
GROUP BY customer_city
ORDER BY total_orders DESC
LIMIT 10;
"""

df = pd.read_sql(query, conn)

plt.figure()
plt.bar(df['customer_city'], df['total_orders'])
plt.xticks(rotation=45)
plt.title("Top Cities by Orders")
plt.show()
```
